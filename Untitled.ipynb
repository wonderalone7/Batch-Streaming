```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Docker PySpark Test")
    .master("local[*]")
    .getOrCreate()
)

print("Spark is working successfully!")
```

```python
data = [
    ("amit", 25),
    ("rahul", 30),
    ("priya", 28)
]

df = spark.createDataFrame(data, ["name", "age"])

df.show()
```
