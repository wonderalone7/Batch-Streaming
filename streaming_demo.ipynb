```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Structured Streaming Demo")
    .master("local[*]")
    .getOrCreate()
)

spark.sparkContext.setLogLevel("ERROR")

print("Streaming Spark Session Created!")
```

```python
streaming_df = (
    spark.readStream
    .format("text")
    .option("maxFilesPerTrigger", 1)
    .load("/app/streaming_input")
)
```

```python
query = (
    streaming_df.writeStream
    .outputMode("append")
    .format("memory")
    .queryName("streaming_table")
    .start()
)
```

```python
spark.sql(
    "SELECT * FROM streaming_table"
).show(truncate=False)
```
